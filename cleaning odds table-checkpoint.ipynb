```python
import pandas as pd

# Replace 'your_file.csv' with the path to your CSV file
file_path = 'big_5_leagues_odds_3-29.csv'
big_df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to confirm it's loaded correctly
print(big_df.head())
```

```python
def decimal_to_american(decimal_odds):
    """
    Convert decimal odds to American odds, adding a '+' sign for positive odds.
    
    Parameters:
    - decimal_odds: float representing the decimal odds.
    
    Returns:
    - str: The American odds as a string, with '+' for positive odds.
    """
    if decimal_odds >= 2.00:
        american_odds = int((decimal_odds - 1) * 100)
    else:
        american_odds = int(-100 / (decimal_odds - 1))
    
    # Add '+' for positive odds
    if american_odds > 0:
        return f'+{american_odds}'
    else:
        return str(american_odds)

# Example usage on a DataFrame column
big_df['Implied Home Odds American'] = big_df['Implied Home Odds'].apply(decimal_to_american)
big_df['Implied Away Odds American'] = big_df['Implied Away Odds'].apply(decimal_to_american)
big_df['Implied Draw Odds American'] = big_df['Implied Draw Odds'].apply(decimal_to_american)
big_df['Draw Odds American'] = big_df['Draw Odds'].apply(decimal_to_american)
big_df['Home Odds American'] = big_df['Home Odds'].apply(decimal_to_american)
big_df['Away Odds American'] = big_df['Away Odds'].apply(decimal_to_american)
# Optionally, display the DataFrame to verify the conversion
big_df[['Implied Home Odds', 'Implied Home Odds American', 
        'Implied Away Odds', 'Implied Away Odds American', 
        'Implied Draw Odds', 'Implied Draw Odds American']]
```

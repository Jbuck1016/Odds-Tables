```python
import pandas as pd
import requests

# Your API key for the odds API
api_key = '6ea2753ab77393e63fe86bb67324fc40'

# Define the endpoint URL
url = 'https://api.the-odds-api.com/v4/sports/'

# Parameters for the API request
params = {
    'apiKey': api_key
}

# Make the GET request
response = requests.get(url, params=params)

# Check if the request was successful
if response.status_code == 200:
    # Extract sports data from the response
    sports_data = response.json()
    
    # Initialize a list to store soccer sports
    soccer_sports = []
    
    # Iterate over the sports data to filter soccer sports
    for sport in sports_data:
        if 'soccer' in sport['key'].lower():
            soccer_sports.append(sport)
    
    # Create a DataFrame from the filtered soccer sports
    soccer_df = pd.DataFrame(soccer_sports)
    print(soccer_df)
else:
    print(f"Error fetching data: {response.status_code}")
```
